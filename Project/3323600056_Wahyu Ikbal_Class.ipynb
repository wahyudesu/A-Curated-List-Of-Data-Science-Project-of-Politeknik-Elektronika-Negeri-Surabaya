```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

class Hero_MLBB:
    
    assest = "3d karakter"
    #Inisialisasi
    def __init__(self, nama, role, level, specialty):
        self.nama = nama
        self.role = role
        self.level = level
        #self.specialty = list(specialty)
        #self.hero = Hero_MLBB(nama, role, level, specialty)
    #Instance method
    def __str__(self):
        return f"Hero: {self.nama}, Role: {self.role}, BP: {self.level}"
    #Method
    def up_level(self, tambahan):
        self.level += tambahan
        print(f"BP {self.nama} bertambah {tambahan} dari 1 menjadi {self.level}")
    def tambah_gambar(self, gambar):
        self.gambar = gambar
    def info(self):
        print(self.nama)
        plt.imshow(mpimg.imread(self.gambar))
        plt.axis('off')
        plt.show()
        
Balmond = Hero_MLBB("Balmond", "Fighter", 1, "Damage Area")
print(Balmond)
Balmond.up_level(2)
```

```python
Balmond.tambah_gambar("balmond.jpg")
Balmond.info()
```

```python
class Item_MLBB:
    
    assest = "2d item"
    
    def __init__(self, nama, harga, fungsi):
        self.nama = nama
        self.harga = harga
        self.fungsi = fungsi
        
    def infokan(self):
        print("Nama: ", self.nama)
        print("Harga: ", self.harga)
        print("Fungsi: ", self.fungsi)
        plt.imshow(mpimg.imread(self.gambar))
        plt.axis('off')
        plt.show()
        
    def tambah_gambar(self, gambar):
        self.gambar = gambar

    def info(self):
        print("Item:", self.nama)
        plt.imshow(mpimg.imread(self.gambar))
        plt.axis('off')
        plt.show()

item = Item_MLBB("BOD", 1000, "Meningkatkan serangan")
item.tambah_gambar("blade of despair.jpg")
item.info()
```
