```python
class Produksi:
    titik1 = 0
    titik2 = 2000
    titik3 = 7000
    
    def __init__(self, produksi):
        self.produksi = produksi
    
    @staticmethod
    def berkurang(*args):
        if len(args) == 0:
            if Produksi.produksi >= Produksi.titik1 and Produksi.produksi <= Produksi.titik2:
                return 1.0
            elif Produksi.produksi > Produksi.titik2 and Produksi.produksi < Produksi.titik3:
                return float(Produksi.titik3 - Produksi.produksi) / (Produksi.titik3 - Produksi.titik2)
            else:
                return 0.0
        else:
            return float(Produksi.titik3 - (args[0] * (Produksi.titik3 - Produksi.titik2)))
    
    @staticmethod
    def bertambah(*args):
        if len(args) == 0:
            if Produksi.produksi > Produksi.titik2 and Produksi.produksi < Produksi.titik3:
                return float(Produksi.produksi - Produksi.titik2) / (Produksi.titik3 - Produksi.titik2)
            elif Produksi.produksi >= Produksi.titik3:
                return 1.0
            else:
                return 0.0
        else:
            return float(Produksi.titik2 + (args[0] * (Produksi.titik3 - Produksi.titik2)))
```
