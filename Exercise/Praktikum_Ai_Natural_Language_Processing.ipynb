```python
# Tambahan
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen(url).read()
soup = BeautifulSoup(html, features="html.parser")

# kill all script and style elements
for script in soup(["script", "style"]):
    script.extract()    # rip it out

# get text
text = soup.get_text()

# break into lines and remove leading and trailing space on each
lines = (line.strip() for line in text.splitlines())
# break multi-headlines into a line each
chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
# drop blank lines
text = '\n'.join(chunk for chunk in chunks if chunk)

print(text)
```

```python
nltk.download('stopwords')

from nltk import word_tokenize

stop_words = set(stopwords.words('indonesian'))
i_filtered_list = [word for word in word_tokenize(text) if not word in stop_words]
i_filtered_list
```

### Tugas Praktikum 2

```python
import nltk
from nltk import CFG
from nltk.parse import RecursiveDescentParser

# Definisi CFG
kalimat = CFG.fromstring("""
    Kalimat -> Subyek Predikat
    Kalimat -> Subyek Predikat Keterangan
    Subyek -> Determinan KataBenda
    Predikat -> KataKerja KataBenda
    Determinan -> 'seorang' | 'seekor'
    KataBenda -> 'manusia' | 'anjing' | 'nasi' | 'daging' | 'air'
    KataKerja -> 'makan' | 'minum'
    Keterangan -> 'kemaren' | 'tadi' """)
rd_parser = RecursiveDescentParser(kalimat)

sentences = [
    "seekor anjing makan daging tadi",
    "seorang manusia minum air kemaren"
]

# Parsing dan menampilkan tree
for sentence in sentences:
    print("Kalimat:", sentence)
    tokens = nltk.word_tokenize(sentence)
    for tree in rd_parser.parse(tokens):
        tree.pretty_print()
```

<div style="border-radius: 10px; border: #0ea5e9 solid; padding: 15px; background-color: #ffffff00; font-size: 100%; text-align: left;">
    <b>üñäÔ∏è Analisis :</b> Sebuah CFG didefisinisakn dengan aturan-atura nproduksi yang menggambarkan sutruktur kalimat yang valid dalam bahasa Indonesia. CFG ini terdiri dari produksi-produski yang mendefinisian hubunan natara subyek, predikat,determinan, kata benda, kata kejra, dan keterangan. Parser rekursif turun diinisialisasi dengn CFG yang telah didefinisikan. Kalimat yang ingin dianalisis kemudian disediakna dalam bentuk list. Setiap kalimat diproses oleh parser, lalu struktru parse  tree ditampilkan oleh output.
